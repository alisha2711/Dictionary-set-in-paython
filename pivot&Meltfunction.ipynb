```python
#Melt function
import pandas as pd
var=pd. DataFrame({"day":[1,2,3,4],"Eng":[22,44,55,33],"Maths" : [33,5,99,88]})
var
```

```python
var=pd. DataFrame({"day":[1,2,3,4],"Eng":[22,44,55,33],"Maths" : [33,5,99,88]})
pd.melt(var)  #reshape pandas function horizontal to vertical 
```

```python
var=pd. DataFrame({"day":[1,2,3,4],"Eng":[22,44,55,33],"Maths" : [33,5,99,88]})
pd.melt(var,id_vars=["day"],var_name="python",value_name="data")
```

```python
#pivot function
var.pivot(index="day",columns="name",values="eng")
```
